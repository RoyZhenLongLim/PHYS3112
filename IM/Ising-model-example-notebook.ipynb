## Experiment 4.1.1: compare a 'cold' start to a 'hot' start

```python
exp1_cold=model.plots(N=30,start='Low',inc=0.05,x0=1,x1=5,T=1)
exp1_cold.mag()
exp1_cold.show()
```

```python
exp1_hot=model.plots(N=30,start='High',inc=0.05,x0=1,x1=5,T=1)
exp1_hot.mag()
exp1_hot.show()
```

For low temperatures, hot start has more variation in magnetization.
This behaviour gradually peters out as temperature rises.

## 4.1.2: On the same size lattice, plot susceptibility versus temperature

## 4.1.3: Maps of spin distribution at four temperatures, on a bigger lattice

```python
for T in [1.5, 2, 3.5, 4]:
    exp3 = model.plots(N=100,start='Low',inc=0.05,x0=1,x1=5,T=T)
    exp3.lattice()
    exp3.show()
```

## 4.1.4: Turn on an external magnetic field and remake the cold start plots for magnetization, magnetic susceptibility and energy, and the four spin maps
### To keep the two terms in the Hamiltonian comparable, B should be on the order of 1; otherwise it doesn't matter (if it's too small) or it dominates (if it's too large).

```python
exp4=model.plots(N=30,start='Low',B=1,inc=0.05,x0=1,x1=5,T=2.5)
exp4.mag()
exp4.show()
exp4.mag_sus()
exp4.show()
exp4.lattice()
exp4.show()
```
