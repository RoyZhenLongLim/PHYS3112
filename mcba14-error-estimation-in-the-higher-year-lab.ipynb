# Higher year lab uncertainties and errors

## Why are uncertainties important?

Uncertainty estimates are fundamental to physics, we aren't just trying to torture you by asking for every number to have an uncertainty. 

Uncertainties are essential for the following:

(1) When comparing experiment with theory, you can't say anything useful unless you know the experimental errors, and the corresponding numbers for the theoretical predictions.

(2) When trying to improve an experiment, we need to know which errors are the most significant, so we can concentrate our efforts on improving those.

## Errors 101: chosing the number of significant figures

If you are meeting a friend for coffee and they ask you "how long will you be?", and you reply, in all seriousness, "9 minutes 43.28672514049 seconds", then your friend would be justified in thinking that there was something wrong with you. A more sensible answer would have been "about 10 minutes".

Similarly, if your doctor asks you for your height in cm, you wouldn't respond with "158.28735281", you would say "158".

In both these everyday examples we have an intuitive feel for how many significant figures are justified. One important aim of the higher year lab is to give you the same intuitive feel for all measurements that you make, and in the results of calculations from those measurements.

If you are unsure of how many significant figures to write down, it is usually acceptable to add one or two more digits, but don't get carried away. To the person marking your physics lab book, too many significant figures looks just as ridiculous as the two examples given above.

## What about the error on the error?

Just as measurements have uncertainties, so do the uncertainties themselves. In first year, we often say "only quote one significant figure in the error", however, this fails in two situations (1) where that significant figure is a "1", in which case you are only specifying the error to a precision of 50%, and (2) if you have done a really careful job of estimating the errors, in which case two, or maybe three, significant figures could be justified. To handle the first situation, use two significant figures if the first one is "1" or "2".

For more information on significant figures, see this https://github.com/phys3112/intro/blob/master/mcba13-writing-numbers.ipynb

## How do you estimate an error in a measurement?

Every measurement has an error (or equivalently, an uncertainty, or tolerance) associated with it. A common refrain from students is "the number was just given, it didn't have an error on it". To which the response is "well, have an educated guess". Such a guess may involve trying to find the datasheet for the sensor that was used. Using "half the smallest division" is often a reasonable guess, although there is no guarantee that it is correct.

In the higher year lab, all we are asking for is that you make a reasonable estimate of the errors in the measurements, and that you briefly comment on how you derived the error.

## Examples of error estimation:

1. A metre ruler, graduated in millimeters. $\pm0.5$mm is reasonable.

2. A 50g brass weight. A quick google search indicates that 1% is a reasonable error for relatively cheap weights. If the exact tolerance is critical, then you should make inquiries about the particular set of weights that you are using.

3. A digital voltmeter. This is tricky since it depends not only on the model of meter, but on the type of measurement and the range you are on. The absolutely accuracy is usually given as a percentage of the full-scale reading of the range, plus a delta in the least significant digit. E.g., for a Fluke 179 multimeter on the DC volt range, the uncertainty is $\pm(0.09\% + 2)$, where the $+2$ is the allowable change in the last digit, so a measurment of 0.500V on the 2V range will have an absolute error of $\pm0.004$V; for AC voltage, and current measurements, the error is between 1% and 1.5%, plus 3.

4. A 10,000 ohm resistor. Resistors come in a wide variety of tolerances. If the resistor has bands of colour on it, one of those will indicate the tolerance. 1% and 5% tolerances are common.

5. A 100 microfarad capacitor. Capacitors often have much looser tolerances than resistors. Large value electrolytic capacitors can easily be 20% or more out, usually on the high side.

## But in first year we were told to do ...

Uncertainty calculations can be made to various degrees of accuracy depending on how critical the knowledge of the error is. In first year Physics we don't want to overwhelm you, so we provide a simple treatment and give you rule-of-thumbs such as "use 1 significant figure in the error" and "error = (max - min) / 2". 

Here is a secret: as in many areas of Physics, what we told you in first year isn't exactly right! 

In 2nd and 3rd year we expect you to perform more sophisticated error analyses. And it doesn't stop there, if you go on to work as a research scientist, you will find that analysing errors can become much more complex again, particularly if you are trying to prove that the speed of light is varying or that one of Einstein's predictions from General Relativity is wrong.

## Common pitfalls

1. Using the full-width-at-half-maximum as an indication of how accurately the position of a peak can be found.

2. Averaging a table of deltas (all the intermediate deltas cancel out).

3. Averaging a series of slopes to find the best fitting straight line.
